```python
import boto3
```

```python
dynamodb = boto3.resource('dynamodb')
jobs_table = dynamodb.Table('jobs')
```

```python
jobs_table.delete_item(Key={'job_id': 'ghactivity_ingest'})
item = {
    'job_id': 'ghactivity_ingest',
    'job_description': 'Ingest ghactivity data to s3',
    'is_active': 'Y',
    'baseline_days': 3
}
jobs_table.put_item(Item=item)
```

```python
jobs_table.delete_item(Key={'job_id': 'ghactivity_transform'})
item = {
    'job_id': 'ghactivity_transform',
    'job_description': 'Transform Files from JSON to Parquet',
    'is_active': 'Y'
}

jobs_table.put_item(Item=item)
```

```python
jobs_table.scan()
```

```python
dynamo_client = boto3.client('dynamodb')
dynamo_client.delete_table(TableName='job_run_details')
```

```python

response = dynamo_client.create_table(
    AttributeDefinitions=[
        {
            'AttributeName': 'job_id',
            'AttributeType': 'S'
        },
                {
            'AttributeName': 'job_run_time',
            'AttributeType': 'N'
        }
    ],
    TableName='job_run_details',
    KeySchema=[
        {
          'AttributeName': 'job_id',
          'KeyType': 'HASH'
        },
        {
          'AttributeName': 'job_run_time',
          'KeyType': 'RANGE'
        },        
    ],
    BillingMode='PAY_PER_REQUEST'
)
```

```python
dynamo_client.list_tables()
```

```python
job_run_details_table = dynamodb.Table('job_run_details')
```

```python
job_run_details_table.scan()
```
